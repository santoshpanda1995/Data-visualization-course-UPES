```python
# create an array

arr=np.array([1,2,3,4,5,6])
print(arr)
```

```python
#Random Matrix Creation

random_matrix=np.random.rand(5,6)
print(random_matrix)
```

```python
#Random Matrix creation with integer values
random_integer=np.random.randint(0,100,size=(5,6))
print(random_integer)
```

```python
#Adding a new row
new_row=np.random.rand(1,6)
print(new_row)
```

```python
random_matrix=np.vstack((random_matrix,new_row))
print(random_matrix)
```

```python
#Slicing

resultant_matrix=random_matrix[:,0:4]
print(resultant_matrix)
```
