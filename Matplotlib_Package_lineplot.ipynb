```python
import matplotlib.pyplot as plt
```

```python
#Line Plot

x=[1,2,3,4,5]
y=[2,4,6,8,10]

#plot
plt.plot(x,y)
plt.title("Basic Line Plot")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.show()
```

```python
#Line Plot with customization

x=[1,2,3,4,5]
y=[2,4,6,8,10]

#plot
plt.plot(x,y,color="red",linestyle="-.",marker="o",markersize=8)
plt.title("Basic Line Plot")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.show()
```

```python
#Multiple lines in a single plot

x=[1,2,3,4,5]
y1=[2,4,6,8,10]
y2=[2,3,5,7,11]

#plot
plt.plot(x,y1,color="red",linestyle="-.",marker="o",markersize=8)
plt.plot(x,y2,color="blue",linestyle="-.",marker="o",markersize=8)
plt.title("Multiple Line Plot")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.show()
```

```python
#Grid lines and ticks

x=[1,2,3,4,5]
y1=[2,4,6,8,10]
y2=[2,3,5,7,11]

#plot
plt.plot(x,y1,color="red",linestyle="-.",marker="o",markersize=8)
plt.plot(x,y2,color="blue",linestyle="-.",marker="o",markersize=8)
plt.title("Multiple Line Plot")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.xticks([0,1,2,3,4,5])
plt.yticks([0,5,10,15,20,25])
plt.grid(True,linestyle="--",alpha=0.6)
plt.show()
```

```python
#subplot

x=[1,2,3,4,5]
y1=[2,4,6,8,10]
y2=[2,3,5,7,11]

plt.subplot(2,2,1)
plt.plot(x,y1,color="red",linestyle="-.",marker="o",markersize=8)
plt.title("First subplot")

plt.subplot(2,2,2)
plt.plot(x,y2,color="blue",linestyle="-.",marker="o",markersize=8)
plt.title("Second subplot")

plt.subplot(2,2,3)
plt.plot(x,y2,color="yellow",linestyle="-.",marker="o",markersize=8)
plt.title("Third subplot")

plt.subplot(2,2,4)
plt.plot(x,y2,color="yellow",linestyle="-.",marker="o",markersize=8)
plt.title("Fourth subplot")



plt.show()
```

```python
#Annotate a point in the figure
plt.style.use("ggplot")
x=[1,2,3,4,5,6,7,8]
y3=[10,8,6,4,4,4,4,4]

plt.plot(x,y3,color="blue",linestyle="-.",marker="o",markersize=8)
plt.title("Annotate Line Plot")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.annotate("Convergence Point",xy=(4,4),xytext=(6,9),arrowprops=dict(facecolor="red"))
plt.show()
```

```python
#Legend

#Multiple lines in a single plot
plt.style.use("ggplot")
x=[1,2,3,4,5]
y1=[2,4,6,8,10]
y2=[2,3,5,7,11]

#plot
plt.plot(x,y1,color="red",label="Linear",linestyle="-.",marker="o",markersize=8)
plt.plot(x,y2,color="blue",label="Non-Linear",linestyle="-.",marker="o",markersize=8)
plt.title("Multiple Line Plot")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.legend(loc="lower right")
plt.show()
```
